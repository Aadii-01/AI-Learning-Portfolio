# **NumPy – Introduction**

## **Q. What is NumPy?**

**NumPy (Numerical Python)** is the fundamental library for numerical computing in Python. It provides a powerful **N-dimensional array (`ndarray`)** object that can efficiently store and process large amounts of data.

Unlike Python lists, NumPy arrays are designed specifically for mathematical and scientific computations.

### NumPy supports:

* 1D Arrays (Vectors)
* 2D Arrays (Matrices)
* 3D Arrays
* Higher-dimensional arrays (4D, 5D, and beyond)

**Example:**

```python
import numpy as np

arr1 = np.array([1, 2, 3])          # 1D Array

arr2 = np.array([[1, 2],
                 [3, 4]])           # 2D Array
```

<br><br>
# Why NumPy Over Python Lists?
**bold text**
Although Python lists are flexible and easy to use, they become inefficient when working with large datasets or performing mathematical operations.

NumPy is specifically optimized for numerical computation.

<br>

## 1. Fixed Data Types (Faster Execution)

Python lists can store different data types together.

```python
my_list = [10, "Hello", 3.14, True]
```

Since every element may have a different type, Python must continuously check the type of each object while processing it.

In contrast, a NumPy array stores elements of the **same data type**.

```python
arr = np.array([10, 20, 30, 40])
```

Because every element has the same type:

* No repeated type checking is required.
* Operations execute much faster.
* Memory layout is predictable.

<br>

## 2. Lower Memory Usage

Python lists store **references (pointers)** to objects rather than the values directly.

```
List

+--------+      +---------+
|Pointer | ---> | Integer |
+--------+      +---------+
|Pointer | ---> | Integer |
+--------+      +---------+
|Pointer | ---> | Integer |
+--------+      +---------+
```

This requires additional memory for every element.

NumPy stores values directly in a continuous block of memory.

```
NumPy Array

+----+----+----+----+
|10  |20  |30  |40  |
+----+----+----+----+
```

**Benefits:**

* Less memory consumption
* Better cache utilization
* Faster data access

<br>

## 3. No Type Checking During Iteration

Suppose we want to add 1 to every element.

Python list:

```python
for x in my_list:
    # Python checks the data type of every element
```

Since every element could have a different type, Python performs type checking repeatedly.

NumPy arrays already know the data type of every element.

```python
arr + 1
```

No repeated type checking is needed, making operations significantly faster.

<br>

## 4. Contiguous Memory Allocation

One of NumPy's biggest advantages is **contiguous memory allocation**.

### Python List

Elements may be scattered across different memory locations.

```
Memory

100 → 5

550 → 10

920 → 15

1400 → 20
```

The list stores pointers to these locations.

To access each element, the CPU must jump between memory addresses.

---
<br>

### NumPy Array

Elements are stored next to each other.

```
Memory

100 → 5

104 → 10

108 → 15

112 → 20
```

The CPU reads sequential memory locations, which is much faster.

---
<br>

# **Benefits of Contiguous Memory**

## SIMD (Single Instruction Multiple Data)

Modern CPUs can perform the same operation on multiple values simultaneously.

Example:

Instead of:

```
1 + 1

2 + 1

3 + 1

4 + 1
```

The CPU can execute:

```
[1 2 3 4] + 1
```

in a single vectorized operation.

This greatly improves performance.

<br>

## Better Cache Utilization

The CPU cache stores nearby memory locations together.

Since NumPy data is contiguous:

* Fewer memory accesses
* Better cache hits
* Faster execution

Python lists often require jumping to different memory locations, resulting in more cache misses.

| Python Lists                                        | NumPy Arrays                                                                                                                                                                                   |
| --------------------------------------------------- | ---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| Can store different data types                      | Stores a single data type                                                                                                                                                                      |
| Higher memory usage                                 | Lower memory usage                                                                                                                                                                             |
| Slower for numerical computations                   | Highly optimized for numerical computations                                                                                                                                                    |
| Dynamic size                                        | Fixed size after creation (can create new resized arrays)                                                                                                                                      |
| Supports insertion, deletion, append, concatenation | Supports most array operations along with powerful mathematical, statistical, linear algebra, broadcasting, vectorization, reshaping, indexing, slicing, and scientific computing capabilities |
| Best for general-purpose programming                | Best for data science, machine learning, AI, scientific computing, and numerical analysis                                                                                                      |

# **Applications of NumPy**

* **Mathematical Computing (MATLAB Replacement):** Fast numerical computations, matrix operations, linear algebra, statistics, and scientific computing.

* **Data Visualization (Matplotlib):** Used to create line plots, bar charts, scatter plots, histograms, and other visualizations.

* **Backend for Python Libraries:** Powers libraries like **Pandas**, **SciPy**, **Scikit-learn**, **OpenCV**, and many others by providing efficient array operations.

* **Digital Image Processing:** Represents images as multidimensional arrays for image filtering, transformation, and analysis.

* **Game Development:** Used for matrix calculations, physics simulations, and grid-based games like Connect Four.

* **Machine Learning & AI:** Forms the foundation for data preprocessing, feature engineering, and mathematical computations used in ML models.


<br>

# **Key Takeaways**

* NumPy is the standard library for numerical computing in Python.
* It provides efficient **N-dimensional arrays**.
* Arrays use **fixed data types**, making computations much faster.
* NumPy stores data in **contiguous memory**, improving CPU cache efficiency.
* Vectorized operations leverage **SIMD**, enabling multiple values to be processed simultaneously.
* Compared to Python lists, NumPy offers **better speed, lower memory usage, and a rich set of mathematical operations**, making it an essential library for AI, Machine Learning, Data Science, and Scientific Computing.

# **Linear Algebra**

````python
a = np.ones((2,3))
b=np.full((3,2),2)
print(a)
print(b)

#Matrix multiplication
np.matmul(a,b)

#Matrix addition
np.add([1,2],[2,2])

#Matrix subtraction
np.subtract([1,2],[3,2])


#transpose
np.transpose(a)

#inverse
np.linalg.inv([[1,2],[2,1]])

#matrix rank
np.linalg.matrix_rank(a)

#eigenvalues and eigenvectors
np.linalg.eig([[1,2],[2,1]])

#Trace sum of diagonal elements
np.trace(a)
````

````python
# Find the determinant
c=np.identity(3)
np.linalg.det(c)
````

````python
# Dot product
np.dot([1,2],[1,2])

#cross product
np.cross([1,2,3],[2,3,4])

#inner product
np.inner([1,2],[2,1])

#outer product
np.outer(a,b)
````

````python
#Solve linear equation
A=np.array([[2,1],[1,-1]])
B=np.array([1,2])
np.linalg.solve(A,B)
````

````python
#Identity matrix
I = np.eye(3)

#Diagonal matrix
np.diag([1,2,3])
````

# **Matplotlib**

Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
x = [0,1,2,3,4,5]
y= [4,5,6,7,8,9]
_y=[1,2,3,4,5,6]

#resizing the graph
plt.figure(figsize=(5,3),dpi=500)

plt.plot(x,y,label='2x', color='grey', linewidth='5', marker='.', markersize=10, markeredgecolor='white',linestyle='dashed')

#shorthand notation
#fmt = '[color][marker][line]
plt.plot(x,_y,'b.--',label='2y')


#Line number two
x2=np.arange(0,4.5,0.8)
plt.plot(x2,x2**2,'r',label='x^2')





plt.title("Pehla graph!", fontdict={'fontname':'serif', 'fontsize':30})
#To change the font details can use fontdict param
plt.xlabel("This is the x axis")
plt.ylabel("This is the y axis")

#to get only integer values on the label
plt.xticks(x)
plt.yticks(y)

#to add a legend on the axis
plt.legend()

#save the graph generated
plt.savefig('pehla_graph.png')


plt.show()
```

# Bar Chart

```python
labels=['a','b','c']
values = [1,6,4]
bar = plt.bar(labels,values,color='green')
bar[0].set_hatch('/')
bar[1].set_hatch('+')
bar[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

# **Real-world examples**

Working with the gas prices csv file

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

## Vertcal bar chart

```python
latest = gas.iloc[-1]

countries = latest.index[1:]
prices = latest.values[1:]

plt.figure(figsize=(12,6))

plt.bar(countries, prices)

plt.title("Gas Prices in Latest Year")

plt.xticks(rotation=45)

plt.show()
```

## Horizontal bar chart

```python
plt.figure(figsize=(10,6))

plt.barh(countries, prices)

plt.title("Gas Prices")

plt.show()
```

## Pie chart

```python
plt.figure(figsize=(8,8))

plt.pie(
    prices,
    labels=countries,
    autopct='%1.1f%%'
)

plt.title("Gas Price Distribution")

plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,6))

plt.hist(
    gas['USA'],
    bins=8
)

plt.title("USA Gas Price Distribution")

plt.show()
```

## Box plot

```python
plt.figure(figsize=(8,6))

plt.boxplot([
    gas['USA'],
    gas['Canada'],
    gas['Germany']
])

plt.xticks(
    [1,2,3],
    ['USA','Canada','Germany']
)

plt.show()
```

## Area chart

```python
plt.figure(figsize=(10,6))

plt.fill_between(
    gas['Year'],
    gas['USA'],
    alpha=0.4
)

plt.plot(gas['Year'], gas['USA'])

plt.show()
```

## Stack plot

```python
plt.figure(figsize=(10,6))

plt.stackplot(
    gas['Year'],
    gas['USA'],
    gas['Canada'],
    gas['Australia'],
    labels=['USA','Canada','Australia']
)

plt.legend()

plt.show()
```

## Stem plot

```python
plt.figure(figsize=(10,6))

plt.stem(
    gas['Year'],
    gas['USA']
)

plt.show()
```

## Step plot

```python
plt.figure(figsize=(10,6))

plt.step(
    gas['Year'],
    gas['USA']
)

plt.show()
```

## Error chart plot

```python
import numpy as np

errors = np.random.rand(len(gas))*0.15

plt.figure(figsize=(10,6))

plt.errorbar(
    gas['Year'],
    gas['USA'],
    yerr=errors,
    fmt='o-'
)

plt.show()
```

## Professional Plot

```python
plt.figure(figsize=(12,8))

for country in gas.columns[1:]:
    plt.plot(
        gas['Year'],
        gas[country],
        marker='o',
        linewidth=2,
        markersize=5,
        label=country
    )

plt.title('Gas Prices by Country', fontsize=20)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Price (USD)', fontsize=15)

plt.grid(True, linestyle='--', alpha=0.5)

plt.legend(
    loc='upper left',
    bbox_to_anchor=(1.02,1)
)

plt.tight_layout()

plt.savefig(
    'Gas_Prices_All_Countries.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

# Adding or removing columns

```python
coffee['price'] = 4.99
coffee.head()
```

```python
import numpy as np

coffee['new_price']= np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
coffee.head()
```

```python
coffee.drop(columns = "price") #things get changed in the copy of df
coffee.drop(columns="price", inplace=True) #things get changed in the original df
```

```python
coffee.head()
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
coffee.head()
```

```python
coffee.rename(columns={'new_price':'price'},inplace=True)
coffee.head()
```

```python
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split().str[0]
bios_new.head()
```

```python
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
bios_new.head()
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new.head()
```

```python
# Saving our updated dataframe
bios_new.to_csv('bios_new.csv', index=False)
```

```python
# Using Lambda & Custom functions w/ .apply()

bios['height_category']=bios['height_cm'].apply(lambda x: 'Short' if x<160 else ('Average' if x<180 else 'Tall'))
bios.head()
```

```python
def categorize_athlete(row):
  if row['height_cm'] <170 and row['weight_kg'] < 70:
    return 'lightweight'
  elif row['height_cm'] <185 and row['weight_kg'] <= 80:
    return 'MiddleWeight'
  return 'heavyweight'

bios['Category']=bios.apply(categorize_athlete,axis=1)
bios.head()
```

# **Handling Null values** (NaNs) | .fillna()  .interpolate() .dropna() .isna() .notna()

```python
coffee.loc[[3,5], 'Units Sold']=np.nan
coffee.isna().sum()
```

```python
coffee['Units Sold'] = coffee['Units Sold'].interpolate()
```

```python

coffee
```

```python
# to drop all the Nas
coffee=coffee.dropna(subset=['Units Sold'], inplace=True)

# to check whether its na or not
coffee[coffee['Units Sold'].isna()]
coffee[coffee['Units Sold'].notna()]
```

# **Aggregating Data**


```python
bios['born_city'].value_counts()
```

```python
bios[bios['born_country']=='USA']['born_region'].value_counts()
```

```python
# Using Groupby- .groupby() .sum() .mean() .agg()
coffee.groupby(['Coffee Type'])['Units Sold'].sum()
coffee.groupby(['Coffee Type'])['Units Sold'].mean()
```

```python
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'price':'mean'})
```

```python
# Pivot table
pivot = coffee.pivot(columns='Coffee Type',index='Day', values='revenue')
pivot
```

```python
pivot.loc['Monday','Latte']
```

```python
pivot.sum()
```

```python
# Group by combined with Datetime operations
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios.groupby(bios['born_date'].dt.year)['name'].count()
```
